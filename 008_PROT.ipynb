# Translating RNA into Protein
## Теория
**Белок** --- полимер аминокислот. В общем случае, существует 20 аминокислот.
Основная структура белка обусловлена последовательностью аминокислоты.
Белок отвечает за большую часть функций внутри клеток, например, ферменты, строительные белки, транспортные и т.д.. 
Однако, некоторые функции все же заложены в другие молекулы.

**Пептид** --- последовательность нескольких аминокислот, промежуточная структура между белком и аминокислотой.

Белок может состоять из одного или нескольких пептидов.

**Кодон** --- последовательность из трех азотистых оснований на мРНК и обозначают аминокислоты.

С помощью кодона можно зашифровать 64 последовательности. 
Из них три (UAA, UAG, UGA) не обозначают никаких аминокислот и являются признаком конца процесса создания белка.
Кодон **AUG** шифрует метионин, а также может обозначать начало процесса создания белка.

*Одну аминокислоту могут шифровать несколько кодонов.*

**Рибосома** --- это органелла, которая в процессе **трансляции** создает белок. 
Для этого ей требуется так называемая **транспортная РНК (тРНК)**.

На одном конце тРНК находится **антикодон** --- последовательность из трех оснований, комплементарных (A <-> U, C <-> G) основаниям кодона на мРНК. 
А на другом конце находится аминокислота.

1. Рибосома движется по мРНК и считывает очередной кодон.
2. Приходит тРНК с комплементарным кодоном.
3. Аминокислота добавляется к растущей пептидной цепи, оставшаяся часть тРНК освобождается из рибосомы.
5. Рибосома переходит к следующему кодону.

**Ген** --- промежуток цепи ДНК, который содержит инструкцию по созданию того или иного белка или РНК.

## Решение задачи

### Комментарий
Библиотека biopython предназначена для работы с биоданными.

unambiguous -- определенный

- unambiguous_rna_by_name - словарь с однозначными таблицами обозначений аминокислот по кодонам из рнк (то есть с урацилом U)
- unambiguous_rna_by_name - словарь с однозначными таблицами обозначений аминокислот по кодонам из днк (то есть с тимином T)

ambiguous -- неопределенный

В таблицах ambiguous есть еще буква N, которая соответствует сразу всем 4 буквам азотистых оснований. 
Это удобно, так как есть четверки кодонов, которые вне зависимости от того, какая буква стоит на определенном месте (заменяем ее N), кодируют одну и ту же аминокислоту.

В словарях есть сразу несколько таблиц, так как генетических код не одинаков у всех организмов. 

Стоп-кодоны лежат в отдельном списке .stop_codons


```python
from Bio.Data import CodonTable

standard_rna = CodonTable.unambiguous_rna_by_name["Standard"].forward_table
stops = CodonTable.unambiguous_rna_by_name["Standard"].stop_codons

# with open("008_prot.txt", "r") as file:
#     s = file.readline()

s = input()

aminoacid_string = ""

for i in range(0, len(s), 3):
    if s[i:i+3] in stops:
        aminoacid_string += "Stop"
    else:
        aminoacid_string += standard_rna[s[i:i+3]]

print(aminoacid_string)
```
