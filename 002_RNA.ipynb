# Transcribing DNA into RNA 

## Теория 
**РНК** --- еще одна нуклеиновая кислота, в которой сахаром является **рибоза**.

Азотистые основания РНК:
- аденин (A);
- цитозин (C);
- гуанин (G);
- урацил (U).

**Транскрипциия** --- процесс, в ходе которое на основе ДНК, как по матрице, составляется РНК.
В результате транскрипции получается один из видов РНК --- **матричная (информационная)**. 
Обозначается мРНК (иРНК). В мРНК азотистые основания совпадают (тимин T заменяется на урацил U).

У эукариот ДНК остается в ядре, а РНК выходит за его пределы, передавая информацию другим частям клетки.

РНК также встречается во всех организмах на Земле.
