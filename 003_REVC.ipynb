# The Secondary and Tertiary Structures of DNA
## Теория
Структура ДНК:
- Молекула состоит из двух цепочек, идущих в противоположных направлениях.
- Азотистое основание одной цепочки связано с азотистым основанием другой по правилу: 

    - Аденин <---> Тимин,
    - Цитозин <---> Гуанин.
    
    Это правило называется **принципом комплементарности**.
- Эти две цепочки переплетены в длинную двойную спираль.

# Решение

```python
with open("003_revc.txt", "r") as file:
    s = file.readline()

s = s[::-1]
s = s.replace("A", "t")
s = s.replace("T", "A")
s = s.replace("t", "T")
s = s.replace("C", "g")
s = s.replace("G", "C")
s = s.replace("g", "G")

with open("003_revc_answer.txt", "w") as file:
    print(s, file=file)
```
