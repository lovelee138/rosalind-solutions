```python
with open("005_gc.txt", "r") as file:
    strings = file.readlines()

gc_info = dict()

number = int(strings[0][10:14])
s = ''

for i in range(1, len(strings)):
    if strings[i][0] == '>':
        s = s.replace('\n', '')
        gc = (s.count('G') + s.count('C'))/len(s)*100
        gc = round(gc, ndigits=4)
        gc_info[number] = gc

        number = int(strings[i][10:14])
        s = ''
    else:
        s += strings[i]

id = max(gc_info, key=gc_info.get)
print("Rosalind_", id, sep='')
print(gc_info[id])
```
