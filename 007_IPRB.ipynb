# Mendel's First Law
## Теория
Теория смешанного наследования имела распространение в 19-ом веке.
Она предполагала, что организм должен сочетать смесь признаков своих родителей.
Однако, она была опровержена эмпирически. 
Например, двое низких особей могут произвести высокое потомство.
Вдобавок к этому, если бы она была верной, все виды стремились бы к некоторому среднему.

Мендель считал, что внешние признаки необходимо делить на дискретные **факторы**. 
Он предположил, что каждый такой фактор может принимать две формы, называемые **аллелями**: доминантную и рецессивную.

**Первый закон Менделя**:
Каждый организм имеет два аллеля заданного фактора. Причем одной копии доминантного аллеля достаточно для проявления этого фактора. 

Это значит, что для проявления рецессивного признака необходимо, чтобы оба аллеля были рецессивны.

Индивид с двумя одинаковыми аллелями **гомозиготен** этому фактору, а с разными **гетерозиготен**.

**Генотип** --- генетический состав.

**Фенотип** --- набор проявляемых физических признаков.

## Решение

### Общий случай

Пусть $k$ -- количество доминантных гомозигот, $m$ -- количество гетерозигот, $n$ -- количество рецессивных гомозигот.

- Количество возможных выборов пар: $$A_{n+k+m}^2$$

Выпишем все возможные комбинации для общего решения подобных задач.

Пусть "A" --- доминтная гомозигота, "B" --- гетерозигота, "C" --- рецессивная гомозигота.

1. "A" + "A"

  Во всех случаях, то есть с вероятностью $1$, дает проявление признака (AA, AA, AA, AA).

  Количество: $k\cdot(k-1)$.

2. "A" + "B"

  Во всех случаях, то есть с вероятностью $1$, дает проявление признака (AA, AA, Aa, Aa).

  Количество: $k\cdot m + m \cdot k = 2km$.

3. "A" + "C"

  Во всех случаях, то есть с вероятностью $1$, дает проявление признака (Aa, Aa, Aa, Aa).

  Количество: $k \cdot n + n \cdot k = 2kn$.

4. "B" + "B"

  Вероятность проявления признака $0.75$ (AA, Aa, aA), отсутствия --- $0.25$ (aa).

  Количество: $m\cdot (m-1)$.

5. "B" + "C"

  Вероятность проявления признака $0.5$ (Aa, Aa), отсутствия --- $0.5$ (aa, aa).

  Количество: $m\cdot n + n \cdot m = 2mn$.

6. "C" + "C"

  Вероятность отсутсвия признака $1$ (aa, aa, aa, aa).

  Количество: $n \cdot (n-1)$.

### Данная задача
Вероятность проявления признакака можно посчитать как $1 - \text{вероятность отсутствия}$.

Вероятность отсутствия:
$$\overline{P} = \frac{m\cdot (m-1)}{(k+m+n)\cdot (k+m+n-1)} \cdot 0.25 + \frac{2mn}{(k+m+n)\cdot (k+m+n-1)} \cdot 0.5 + \frac{n\cdot (n-1)}{(k+m+n)\cdot (k+m+n-1)} \cdot 1$$

Вероятность проявления:
$$P =1 - \left( \frac{m\cdot (m-1)}{(k+m+n)\cdot (k+m+n-1)} \cdot 0.25 + \frac{2mn}{(k+m+n)\cdot (k+m+n-1)} \cdot 0.5 + \frac{n\cdot (n-1)}{(k+m+n)\cdot (k+m+n-1)} \cdot 1\right)$$

```python
def prb_dominant(k, m, n):
    prob = 1 - m*(m-1)/((k+m+n)*(k+m+n-1))*0.25  - n*(n-1)/((k+m+n)*(k+m+n-1))*1 - 2*m*n/((k+m+n)*(k+m+n-1))*0.5
    return prob
```

```python
k = 19
m = 29
n = 22

print(prb_dominant(k, m, n))
```
