# Counting Point Mutations
## Теория
Мутация --- это ошибка при процессе копирования генов.

Виды мутаций:
- **Удаление** части цепочки,
- **Дублирование** части цепочки,
- **Инверсия** части цепочки,
- **Вставка** части цепочки,
- **Транслокация** двух частей цепочек.

Мутация может нести и положительный эффект. Так, большие изменения в ходе эволюции --- это, обычно, совокупность многих мелких положительных мутаций.

**Точечная мутация** --- мутация, которая меняет одно азотистое основание на другое. В случае ДНК --- меняет сразу всю пару (в силу принципа комплементарности).

Если ДНК-цепочки двух различных организмов унаследованы от одной цепочки-предка, то они называются **гомологичными**. 
Количество различий (**расстояние Хэмминга**) между ними показывает *минимальное число точечных мутаций*. Причем это число не обязательно достигается, так как мутация могла произойти и вернуться обратно.

**Принцип Парсимонии**: самое простое объяснение из всех возможных является самым верным.
Из принципа парсимонии интересно минимальное число точечных мутаций, которое могло произойти.
Оно как раз равно расстоянию Хэмминга между двумя цепочками.


## Решение

````python
with open("006_hamm.txt", "r") as file:
    s = file.readline()
    t = file.readline()
````

````python
def hamming_distance(s, t):
    if len(s) != len(t):
        return -1

    count = 0
    for i in range(len(s)):
        if s[i] != t[i]:
            count += 1
    
    return count
````

````python
print(hamming_distance(s, t))
````

## Другое решение

````python
def hamming_distance_2(s, t):
    return sum(x[0] != x[1] for x in zip(s, t))

s = "ACGGTCG"
t = "CCGCTAG"

print(hamming_distance_2(s, t))
````
