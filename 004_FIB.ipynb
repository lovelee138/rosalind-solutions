# Rabbits and Recurrence Relations
Задача (1202, Леонардо Фибоначчи)

Предположим следующее:
1. Популяция кроликов начинается с одной новорожденной пары.
2. Кролики достигают репродуктивного возраста через месяц
3. Двое кроликов, достигших репродуктивного возраста производят за 1 месяц k пар кроликов.
4. Кролики не прекращают размножаться и не умирают.

Сколько пар кроликов будет на n-ый месяц?

## Решение
### Случай k=1
На первый и второй месяц будет по одной паре кроликов. 
Так как на второй месяц кролики стали репродуктивного возраста, то на третий месяц их станет 2 пары: 1 исходная пара и 1 новорожденная.

На четвертый месяц пар кроликов станет 3: 2 пары с третьего месяца (1 взрослая и 1 детская, обе стали взрослыми), 1 пара детская.

На пятый месяц пар кроликов станет 5: 3 пары с четвертого месяца (некоторые были детскими, теперь все стали взрослыми) и еще 2 пары с третьего месяца (они стали взрослыми на четвертый месяц) дают в качестве потомства по одной новой паре.

Таким образом, число пар с течением времени --- классическая последовательность Фибоначчи:
$$F_{n} = F_{n-1}+F_{n-2},$$
где $F_1=F_2 = 1$

### Случай $k!=1$
В этом случае в формуле добавляется коэффициент, на который нужно домножить количество пар, которое было за два месяца до текущего. Именно эти пары дают потомство, но теперь не по одной новой паре, а по k пар.

Тогда общая формула выглядит так:
$$ F_{n} = F_{n-1} + k \cdot F_{n-2}.$$

```python
def fibb(n, k):
    if n == 1:
        return 1
    if n == 2:
        return 1
    
    first = 1
    second = 1
    for i in range(3, n+1):
        fib_n = second + first*k
        first = second
        second = fib_n
    return fib_n
```

```python
print(fibb(33, 2))
print(fibb(33, 1))
```
