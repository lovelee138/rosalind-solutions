# Counting DNA Nucleotids

## Теория
Большинство **эукариотических** клеток содержит ядро, но не все. 
Например, зрелые эритроциты млекопитающих освобождают больше места для гемоглобина, избавлясь от **ядра**.
Бывают также клетки с несколькими ядрами.

Большая часть ядра плотно заполнена веществом, состоящим из многих макромолекул и называемым **хроматином**.

Митоз - процесс деления клеток. В процессе митоза хроматин уплотняется и вытягивается в длинные тонкие нити -- **хромосомы**.

Один из видом макромолекул, содержащихся в хроматине, -- **нуклеиновая кислота**. В начале XX века ученые пришли к следующему выводу: 
нуклеиновые кислоты являются **полимерами**, то есть цепочками повторяющихся много раз простых последовательностей -- **мономеров**. 
Мономер нуклеиновой кислоты называется **нуклеотидом** и является единичным отрезком длины цепочки. 

Нуклеотид состоит из трех частей:
- молекула **сахара**,
- отрицательно заряженный ион **фосфат**,
- **азотистое основание** -- химическое соединение.

Полимер формируется за счет того, что сахар одного нуклеотида и соединяется с фосфатом другого, образуя **сахарофосфатный остов**.
Фишка в том, что сахар и фосфат в рамках одной нуклеиновой кислоты остаются одними и теми же. Поэтому разные цепочки (нити) кислот 
определяются порядком азотистых оснований.

Например, **ДНК -- дезоксирибонуклеиновая кислота**. 
Это кислота с сахаром "дезоксирибоза". Азотистые основания у нее: 
- аденин (A), 
- цитозин (C), 
- гуанин (G), 
- тимин (T).

Был обнаружен у всех организмов, даже у вирусов, которых не всегда к живым-то причисляют.


## Задача
Для данной последовательности ДНК выписать количество символов A, C, G, T, встречающихся в заданной строке

```python
with open("001_dna.txt", "r") as file:
    s = file.readline()
    print(s.count('A'), s.count('C'), s.count('G'), s.count('T'))
```
